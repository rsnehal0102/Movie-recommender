Importing the libraries and neccessary cosine_simalirty function required

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data loading


```python
movies_data = pd.read_csv('/content/movies.csv')
```

Sample 5 rows printing

```python
movies_data.head()
```

```python
movies_data.shape # printing the number of rows and columns
```

Selecting the required features

```python
selected_features = ['genres','keywords','overview','cast','director']
print(selected_features)
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['overview']+' '+movies_data['cast']+' '+movies_data['director']
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

**COSINE SIMILARITY**

```python
similarity = cosine_similarity(feature_vectors)
print(similarity) 
```

Interaction with user

```python
movie_name = input(' Please type your favourite movie name for which you want similar movies to be presented: ')
```

```python
list_of_all_titles = movies_data['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = find_close_match[0]
print(close_match)
```

```python
#Getting the index to the most related movie
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
print('Movies suggested for according to the given input movie(top 10 most similar): \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<=10):
    print(i, '.',title_from_index)
    i+=1
```
